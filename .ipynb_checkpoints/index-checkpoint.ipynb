```python
# 상위 4개 DevType을 추출
top_devtypes = df['DevType'].value_counts().head(4).index

# DevType을 기준으로 묶기 (상위 4개만 선택)
grouped_by_devtype = df[df['DevType'].isin(top_devtypes)].groupby('DevType')

# 색상 리스트 정의
colors = ['blue', 'green', 'orange', 'red', 'purple']

# 각 DevType에 대한 경력 분포 시각화
plt.figure(figsize=(15, 8))
for i, (devtype, data) in enumerate(grouped_by_devtype):
    sns.histplot(data=data, x='WorkExp', label=devtype, kde=True, bins=20, alpha=0.5, color=colors[i])

plt.title('상위 4개의 개발 유형별 경력 분포', fontsize=16)
plt.xlabel('경력')
plt.ylabel('빈도수')
plt.legend(title='DevType', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.show()
```

```python
df['CompTotal'].isna().sum()
```

```python
# 최빈값 계산
mode_value = df['CompTotal'].mode()[0]

# 결측치 대체
df['CompTotal'].fillna(mode_value, inplace=True)
```

```python
# 환율 적용
exchange_rate = 1296.38
df['CompTotal_in_krw'] = df['CompTotal'] * exchange_rate
df['CompTotal_in_krw']
```

```python
# DevType별로 그룹화하고 각 그룹의 CompTotal_in_krw 평균값 계산
grouped_by_devtype = df.groupby('DevType', as_index=False) \
                       .agg(mean_sal=('CompTotal_in_krw', 'mean')) \
                       .sort_values('mean_sal', ascending=False)
grouped_by_devtype
```

```python
grouped_by_devtype = grouped_by_devtype[grouped_by_devtype['DevType'] != 'Blockchain']
grouped_by_devtype
```

```python
pd.set_option('display.float_format', lambda x: '%.6f' % x)
grouped_by_devtype.query('DevType == "Developer, back-end"')['mean_sal']
```

```python
# 로그 스케일로 변환
grouped_by_devtype['log_mean_sal'] = np.log(grouped_by_devtype['mean_sal'])

# 상위 5개에서 10개의 DevType만 선택
top_devtypes_subset = grouped_by_devtype.iloc[0:10]

# 그래프 크기 설정
plt.figure(figsize=(12, 6))

# 막대 그래프 그리기
sns.barplot(x='log_mean_sal', y='DevType', data=top_devtypes_subset, palette='viridis')

# 그래프 제목 설정
plt.title('DevType별 CompTotal_in_krw 평균값 (상위 5~10) - 로그 스케일', fontsize=16)

# x축 레이블 설정
plt.xlabel('CompTotal_in_krw 평균값 (로그 스케일)')

# y축 레이블 설정
plt.ylabel('DevType')

# 그래프 표시
plt.show()
```
